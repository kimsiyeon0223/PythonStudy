## **컬렉션 타입**
### : 다양한 데이터를 그룹화하여 관리할 수 있는 자료형
<br />

###**List** : 여러 아이템을 순서대로 저장, 아이템 변경 가능, 중복된 값의 저장이 허용
### **Tuple** : 여러 아이템을 순서대로 저장, 아이템 변경 불가능, 중복된 값의 저장이 허용
### **Set**: 여러 아이템을 순서 없이 저장, 아이템 변경 가능, 중복된 값의 저장이 허용 불가
### **Dictionary**: 키-값 쌍의 아이템을 저장, 아이템 변경 가능, 키(중복X), 값(중복O)

##< List 자료형 >
### : 대괄호[]를 사용하여 생성, 각 요소는 쉼표(,)로 구분
<br />

## 1. 리스트 생성
### 리스트명 = [요소1, 요소2, ...]
```python
list1 = [] # 빈 리스트
list2 = [1, 2, 3] # 정수형 리스트
list3 = ['a', 'b', 'c'] # 문자형 리스트
list4 = [1, 'a', 2, 'b'] # 다양한 자료형의 혼합
list5 = [1, 2, [3, 4]] # 리스트 안에 리스트 (중첩 리스트)
```
<br />

## 2. 리스트 인덱싱 & 슬라이싱
### 리스트 인덱싱
```python
list = [1, 2, 3, 4, 5]
print(list[0]) # 출력: 1
print(list[-2]) # 출력: 4

# 리스트 슬라이싱
print(list[1:3]) # 출력: [2, 3]
```
<br />

## 3. 리스트 연산
```python
# 3-1. 리스트 합치기 ( + )
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list3 = list1 + list2
print(list3) # 출력: [1, 2, 3, 4, 5, 6]

# 3-2. 리스트 반복하기 ( * )
print(list1 * 3) # 출력: [1, 2, 3, 1, 2, 3, 1, 2, 3]

# 3-3. 리스트 길이 구하기: len()
print(len(list1)) # 출력: 3
```
<br />

## 4. 리스트 수정, 삭제 및 추가
```python
# 4-1. 리스트 요소 수정
list = [1, 2, 3] # [1, 2, 4]
list[2] = 4
print(list) # 출력: [1, 2, 4]

# 4-2. 리스트 요소 삭제: del(delete의 축약어)
del list[1]
print(list) # 출력: [1, 4]

# 4-3. 리스트 요소 추가: append()
list.append(5) # 제일 마지막 위치에 요소를 추가
print(list) # 출력: [1, 4, 5]
```
<br /> <br />

##< 튜플(Tuple)자료형 >
###: 순서 O, 변경 X, 중복된 값의 저장 O
###: !!한 번 생성된 튜플은 변경이 불가능!!
###: 소괄호 ()를 사용하여 생성하고, 각 요소는 쉼표(,)로 구분
<br />

## 1. Tuple 생성
```python
tuple1 = () # 빈 튜플
tuple2 = (1, 2, 3) # 정수형 튜플
tuple3 = ('a', 'b', 'c') # 문자형 튜플
tuple4 = (1, 'a', 2.0, 'b') # 다양한 자료형의 혼합
tuple5 = (1, 2, (3, 4)) # 튜플 안에 튜플 (중첩 튜플)

# 단일 요소(1개의 값)를 가지는 튜플을 생성할 때는 요소 뒤에 쉼표를 반드시! 붙여야 한다.
sigle_element_tuple = (1,)
```
<br />

## 2. Tuple 인덱싱 & 슬라이싱
```python
tuple = (1, 2, 3, 4, 5)
print(tuple[1]) # 출력: 2
print(tuple[-1]) # 출력: 5

print(tuple[1:3]) # 출력: (2, 3)
```
<br />

## 3. Tuple 연산
```python
# 3-1. Tuple 합치기 ( + )
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
print(tuple1 + tuple2) # 출력: (1, 2, 3, 4, 5, 6)
# 3-2. Tuple 반복하기 ( * )
print(tuple1 * 2) # 출력: (1, 2, 3, 1, 2, 3)
```
<br />

## 4. Tuple 관련 함수(내장 함수)
```python
# 4-1. 요소의 위치를 반환하는 메서드: index()
tuple = (1, 2, 3, 4)
print(tuple.index(3)) # 출력: 2

# 4-2. 주어진 값과 일치하는 요소의 개수를 반환하는 메서드: count()
tuple = (1, 2, 3, 2, 2, 3, 4, 2)
print(tuple.count(2)) # 출력: 4

# 튜플은 요솟값을 변경할 수 없기 때문에 값을 변경하는 내장 함수가 없다.
```
